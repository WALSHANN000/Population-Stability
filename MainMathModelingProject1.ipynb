```python
import numpy as np
import matplotlib.pyplot as plt
```

In this project I'm modeling a plant (X) and prey(Y) populations and looking at the impact a predator population (Z) will have on the other two

populations are measured in thousands

```python
Dt=.001 #time step size (in years)
X_start=100 #starting plant population
Y_start=10 #starting prey population
Z=.1 #predetor population (using a constant to make the system easier to analyze)
t_start=0
t_end=50
n_steps=int(round((t_end-t_start)/Dt)) #number of time steps in time period
```

```python
#initalizing arrays
X_arr=np.zeros(n_steps+1)
Y_arr=np.zeros(n_steps+1)
dX_arr=np.zeros(n_steps+1)
dY_arr=np.zeros(n_steps+1)
t_arr=np.zeros(n_steps+1)
t_arr[0]=t_start
X_arr[0]=X_start
Y_arr[0]=Y_start
dX_arr[0]=0
dY_arr[0]=0
```

```python
#creating population (X-arr and Y_arr) and derivative population (Dx_att and Dy_arr) arrays
for i in range(1,n_steps+1):
  X=X_arr[i-1]
  Y=Y_arr[i-1]
  t=t_arr[i-1]
  dXdt=.3*X-.015*X*Y #rate of change of plant population: plant growth -plant deaths
  dYdt=.015*X*Y-.2*Y-5*Z*Y #rate of change of prey population: prey growth -natural prey deaths -predatory deaths
  X_arr[i]=X+Dt*dXdt
  Y_arr[i]=Y+Dt*dYdt
  dX_arr[i]=dXdt
  dY_arr[i]=dYdt
  t_arr[i]=t+Dt
dX_arr[0]=dX_arr[1]
dY_arr[0]=dY_arr[1]
```

```python
#population vs time graph
fig=plt.figure()
plt.plot(t_arr,X_arr,linewidth=4,label='X')
plt.plot(t_arr,Y_arr,linewidth=4,label='Y')
plt.title('Population (in thousands) vs Time',fontsize=12)
plt.xlabel('t (in years)',fontsize=12)
plt.ylabel('X(t),Y(t)',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.axis([t_start,t_end,0,150])
plt.legend()
plt.show
```

```python
#population vs population graph
X_equil=13.333+333.333*Z
fig=plt.figure()
plt.plot(X_arr,Y_arr,linewidth=4)
plt.scatter(X_equil,20) #equilibria of the population system- only the x-coordinate is dependent on the predator population
plt.title('Populations (in thousands)',fontsize=12)
plt.xlabel('X pop',fontsize=12)
plt.ylabel('Y pop',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.axis([0,150,0,120])
plt.legend()
plt.show
```

```python
#derivative population vs time graph
fig=plt.figure()
plt.plot(t_arr,dX_arr,linewidth=4,label='dX')
plt.plot(t_arr,dY_arr,linewidth=4,label='dY')
plt.title('Pop Growth Rate',fontsize=12)
plt.xlabel('t (in years)',fontsize=12)
plt.ylabel('dX(t),dY(t)',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.axis([t_start,t_end,-50,50])
plt.legend()
plt.show
```

I found that a smaller predator population will result in lower minimum plant and prey populations and longer time intervals between population peaks.
